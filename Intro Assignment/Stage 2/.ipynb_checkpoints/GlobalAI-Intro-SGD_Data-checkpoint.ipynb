```python
# get tickers from df_sdg
tickers = df_sdg['Ticker'].unique().tolist()
len(tickers)
```

```python
# get price
start = '2017-01-03'
df_price = yf.download(tickers, start)['Adj Close']
```

```python
# clear delisted symbles and symbles with na value
df_price = df_price.dropna(axis=1)
df_price
```

```python
# 542 remains in tickers
tickers = df_price.columns
len(tickers)
```

```python
symbol = yf.Ticker("msft").info
symbol["trailingPE"]
```

```python
symbol["priceToBook"]
```

```python
dict_PE = {}
dict_PB = {}

for company in tickers:
    symbol = yf.Ticker(company).info
    if "forwardPE" in symbol:
        dict_PE[company]= symbol["forwardPE"]
    if "priceToBook" in symbol:
        dict_PB[company]= symbol["priceToBook"]
```
