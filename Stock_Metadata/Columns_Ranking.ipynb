## Check Stock Type Matches

```python
df_type = df[['Name_197', 'stockType_197', 'Name-EOD', 'Type-EOD', 'securityType-FIGI', 'securityType2-FIGI']]

df_type
```

```python
df_type.to_csv('./Data/stockType_table.csv')
```

```python
df_type['stockType_197'] = df_type['stockType_197'].map({'COMMON': 'Common Stock','PREFERRED': 'Preferred Stock'}).fillna(df_type['stockType_197'])
df_type['Type-EOD'] = df_type['Type-EOD'].map({'FUND': 'Mutual Fund'}).fillna(df_type['Type-EOD'])

for col in df_type.columns:
    df_type[col] = df_type[col].str.title()



df_type.head(10)
```

```python
def testMatch(x):
    if x['stockType_197'] == x['Type-EOD'] or (pd.isna(x['stockType_197']) or pd.isna(x['Type-EOD'])):
        if (x['stockType_197'] == x['securityType-FIGI'] or x['stockType_197'] == x['securityType2-FIGI']) or (pd.isna(x['stockType_197']) or pd.isna(x['securityType-FIGI'])):
            return 1
    return -1

df_type['type_match'] = df_type.apply(testMatch, axis=1)
df_type
```

```python
df_type
```

```python
df_mismatch = df_type[df_type['type_match']==-1]

df_mismatch.loc[(df_mismatch['stockType_197']=='Closed-End Fund') & (df_mismatch['Type-EOD']=='Mutual Fund') 
                & (df_mismatch['securityType-FIGI']=='Closed-End Fund'),'type_match'] = 1

df_mismatch.loc[(df_mismatch['stockType_197']=='ETF') & (df_mismatch['Type-EOD']=='ETF') 
                & (df_mismatch['securityType-FIGI']=='ETP'),'type_match'] = 1


df_mismatch = df_mismatch[df_mismatch['type_match']==-1]

df_mismatch
```
