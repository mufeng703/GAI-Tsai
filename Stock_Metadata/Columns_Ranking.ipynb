```python
import numpy as np  
import pandas as pd 
```

```python
path = "./Data/withFIGI.csv"

df = pd.read_csv(path, index_col=0)  
df
```

```python
df = df.drop(['Not_Match', 'Source_Available'], axis=1)

df
```

Country, Sector/Industry, Exchange, ISIN, FIGI

```python
df_ck = df.copy()
df_ck['Ranking'] = 0

df_ck.loc[~df_ck['Country-EOD'].isnull(),'Ranking'] = 1 + df_ck['Ranking'].astype(int)
df_ck.loc[~df_ck['Exchange-EOD'].isnull(),'Ranking'] = 1 + df_ck['Ranking'].astype(int)
df_ck.loc[~df_ck['ISIN'].isnull(),'Ranking'] = 1 + df_ck['Ranking'].astype(int)
df_ck.loc[~df_ck['country-Inv'].isnull(),'Ranking'] = 1 + df_ck['Ranking'].astype(int)
df_ck.loc[~df_ck['Country-IB'].isnull(),'Ranking'] = 1 + df_ck['Ranking'].astype(int)
df_ck.loc[~df_ck['Exchange-IB'].isnull(),'Ranking'] = 1 + df_ck['Ranking'].astype(int)
df_ck.loc[~df_ck['figi'].isnull(),'Ranking'] = 1 + df_ck['Ranking'].astype(int)
df_ck.loc[~df_ck['exchCode-FIGI'].isnull(),'Ranking'] = 1 + df_ck['Ranking'].astype(int)

df_ck[['Name_197','Ranking']]
```
