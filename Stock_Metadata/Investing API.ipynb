```python
import investpy
import numpy as np
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
path = "./Data/final_investpy.csv"

df_inv = pd.read_csv(path, index_col=0)  
# df_inv = investpy.stocks.get_stocks(country=None)
df_inv
```

```python
path = "./Data/LIst of 19.7 Stocks with FIGI if available - By Evelyn.xlsx"

df_stockList = pd.read_excel(path, index_col=0).drop_duplicates()
df_stockList
```

```python
df = df_inv[['full_name', 'name', 'symbol' ]]
df = df.drop_duplicates()
df
```

```python
df_197 = df_stockList[['symbol_x', 'longName']]
df_197
```

```python
df_197[df_197['symbol_x'].isin(df['symbol'])]
```

```python
match = []
allD = []

for index, row in df_197.iterrows():
    
    if row['symbol_x'] in df['symbol'].values :
        df_sym = df[df['symbol'] == row['symbol_x']].copy()
        for index2, row2 in df_sym.iterrows():
            # calculate nlp scores
            fuzz_socre1 = fuzz.token_sort_ratio(row['longName'],row2['full_name'])
            fuzz_socre2 = fuzz.token_sort_ratio(row['longName'],row2['name'])
            
            # append to allD
            row2['Name_197'] = row['longName']
            row2['fuzz_socre1'] = fuzz_socre1
            row2['fuzz_socre2'] = fuzz_socre2
            allD.append(row2)
            
            # get matched result
            if fuzz_socre1>90 or fuzz_socre2>90:
                match.append(row2)

                if len(match) % (len(df_197)//10) == 0:
                    print("progress: %.2f" % (100*len(match)/len(df_197)) + "%")
```

```python
df_match = pd.DataFrame(match)
df_match
```

```python
df_all = pd.DataFrame(allD)
df_all
```

```python
df_all.to_csv('inv_matching_result.csv')
df_match.to_csv('inv_matching_result(matched only).csv')
```
