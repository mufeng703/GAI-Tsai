```python
import investpy
import numpy as np
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
path = "./Data/final_investpy.csv"

df_inv_all = pd.read_csv(path, index_col=0)  
# df_inv_all = investpy.stocks.get_stocks(country=None)
df_inv_all
```

```python
path = "./Data/LIst of 19.7 Stocks with FIGI if available - By Evelyn.xlsx"

df_stockList = pd.read_excel(path, index_col=0)
df_stockList
```

```python
df_inv = df_inv_all[['full_name', 'name', 'symbol' ]]
df_inv = df_inv.drop_duplicates()
df_inv
```

```python
df_197 = df_stockList[['symbol_x', 'longName']]
df_197
```

## Delete 10 most common words in company name

```python
# get 30 most common words (which can not be used in name matching)
from collections import Counter

all_names = pd.concat([df_197['longName'], df_inv['full_name']])
all_names = pd.concat([all_names, df_inv['name']])
names_freq = Counter()
for name in all_names:
    names_freq.update(str(name).split(" "))
key_words = [word for (word,_) in names_freq.most_common(10)]
print(key_words)
```

```python
# delete the 30 most common words in company names
for key in key_words:
    df_197['longName'] = df_197['longName'].str.replace(key, '').copy()
    df_inv['full_name'] = df_inv['full_name'].str.replace(key, '').copy()
    df_inv['name'] = df_inv['name'].str.replace(key, '').copy()

df_197
```

## Matching by symbol and company name

```python
match = []
allD = []

for index, row in df_197.iterrows():
    
    if row['symbol_x'] in df_inv['symbol'].values :
        df_sym = df_inv[df_inv['symbol'] == row['symbol_x']].copy()
        for index2, row2 in df_sym.iterrows():
            # calculate nlp scores
            fuzz_socre1 = fuzz.token_sort_ratio(row['longName'],row2['full_name'])
            fuzz_socre2 = fuzz.token_sort_ratio(row['longName'],row2['name'])
            
            # append to allD
            row2['Name_197'] = row['longName']
            row2['fuzz_socre1'] = fuzz_socre1
            row2['fuzz_socre2'] = fuzz_socre2
            allD.append(row2)
            
            # get matched result
            if fuzz_socre1>90 or fuzz_socre2>90:
                match.append(row2)

    if index % (len(df_197)//10) == 0:
        print("progress: %.2f" % (100*index/len(df_197)) + "%")
```

```python
df_match = pd.DataFrame(match)
df_match
```
