```python
import numpy as np
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
path1 = "./Data/Investing_IB_197.csv"

df_isin = pd.read_csv(path1)  
df_isin
```

```python
path = "./Data/LIst of 19.7 Stocks with FIGI if available - By Evelyn.xlsx"

df_stockList = pd.read_excel(path, index_col=0)  
df_stockList
```

## Data Engineering

```python
Col_inv = ['country-Inv', 'name-Inv', 'full_name-Inv', 'currency-Inv', 'symbol-Inv', 'asset_class-Inv', 'Name-Inv', 'Ticker-Inv', 'Exchange-Inv', 'Type-Inv', 'Market-Inv', 'CUSIP-Inv']
df_inv = df_isin[Col_inv]
df_inv['ISIN'] = df_isin[['ISIN']]

# combin data from web scraping and API
df_inv = df_inv.assign(**{'name-Inv': df_inv['name-Inv'].fillna(df_inv['Name-Inv'])})
df_inv = df_inv.assign(**{'symbol-Inv': df_inv['symbol-Inv'].fillna(df_inv['Ticker-Inv'])})
df_inv.dropna(subset=['name-Inv'], how='all', inplace=True)
df_inv.dropna(subset=['symbol-Inv'], how='all', inplace=True)

# drop columns from web scraping
df_inv = df_inv.drop(columns=['Name-Inv', 'Ticker-Inv', 'Exchange-Inv', 'Type-Inv', 'Market-Inv', 'CUSIP-Inv'])

df_inv
```

```python
df_IB = df_isin[df_isin.columns.drop(Col_inv)]
df_IB.dropna(subset=['Symbol-IB', 'Ticker-IB'], how='all', inplace=True)

# combin data from my method and from Abhir
df_IB = df_IB.assign(**{'Symbol-IB': df_IB['Symbol-IB'].fillna(df_IB['Ticker-IB'])})
df_IB = df_IB.assign(**{'longName-IB': df_IB['longName-IB'].fillna(df_IB['Name-IB'])})
df_IB.dropna(subset=['Symbol-IB'], how='all', inplace=True)
df_IB.dropna(subset=['longName-IB'], how='all', inplace=True)

# drop dupicate columns from my method
df_IB = df_IB.drop(columns=['Ticker-IB', 'Name-IB'])

df_IB
```

```python
df_IB = df_isin[df_isin.columns.drop(Col_inv)]
df_IB.dropna(subset=['Symbol-IB', 'Ticker-IB'], how='all', inplace=True)
df_IB
```

## Matching

```python
df_197 = df_stockList[['symbol_x', 'longName']]
df_197
```

## Matching Investing.com & 19.7k

```python
match = []
allD = []

for index, row in df_197.iterrows():
    
    if row['symbol_x'] in df_inv['symbol-Inv'].values :
        df_sym = df_inv[df_inv['symbol-Inv'] == row['symbol_x']].copy()
        for index2, row2 in df_sym.iterrows():
            # calculate nlp scores
            fuzz_socre1 = fuzz.token_sort_ratio(row['longName'],row2['full_name-Inv'])
            fuzz_socre2 = fuzz.token_sort_ratio(row['longName'],row2['name-Inv'])
            
            # append to allD
            row2['Name_197'] = row['longName']
            row2['fuzz_socre1'] = fuzz_socre1
            row2['fuzz_socre2'] = fuzz_socre2
            allD.append(row2)
            
            # get matched result
            if fuzz_socre1>90 or fuzz_socre2>90:
                match.append(row2)

    if index % (len(df_197)//10) == 0:
        print("progress: %.2f" % (100*index/len(df_197)) + "%")
```

```python
df_match_Inv = pd.DataFrame(match)
df_match_Inv
```

```python
df_match_Inv.to_csv('./Data/metaData_Inv.csv', index=False)
```

## Matching IB & 19.7k

```python
match2 = []
allD2 = []

for index, row in df_197.iterrows()
    
    if row['symbol_x'] in df_IB['Symbol-IB'].values :
        df_sym = df_IB[df_IB['Symbol-IB'] == row['symbol_x']].copy()
        for index2, row2 in df_sym.iterrows():
            # calculate nlp scores
            fuzz_socre1 = fuzz.token_sort_ratio(row['longName'],row2['longName-IB'])
            
            # append to allD
            row2['Name_197'] = row['longName']
            row2['fuzz_socre1'] = fuzz_socre1
            allD2.append(row2)
            
            # get matched result
            if fuzz_socre1>90:
                match2.append(row2)

    if index % (len(df_197)//10) == 0:
        print("progress: %.2f" % (100*index/len(df_197)) + "%")
```

```python
df_match_IB = pd.DataFrame(match2)
df_match_IB
```

```python
df_match_IB.to_csv('./Data/metaData_IB.csv', index=False)
```

```python
df_all = pd.merge(df_match_Inv, df_match_IB, how='outer', on='Name_197')
df_all = df_all.assign(**{'ISIN_x': df_all['ISIN_x'].fillna(df_all['ISIN_y'])})
df_all = df_all.drop(columns=['fuzz_socre1_x', 'fuzz_socre1_y', 'fuzz_socre2', 'ISIN_y'])
df_all = df_all.rename(columns={"ISIN_x": "ISIN"})

df_all
```

```python
df_all.to_csv('./Data/metaData_IB_Inv.csv', index=False)
```
