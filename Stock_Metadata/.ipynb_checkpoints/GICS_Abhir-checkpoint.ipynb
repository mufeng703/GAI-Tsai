```python
for i in symbol_exchange:
    try:
        newdata=f'https://eodhistoricaldata.com/api/fundamentals/{i}?api_token=5f3afd582bd7b4.95720069'
        a = requests.get(newdata).json()
        train = pd.DataFrame.from_dict(a, orient='index')
        train.reset_index(level=0, inplace=True)
        train = train.iloc[0]
        new = train[columns].copy().to_frame().T
        df = pd.concat([df,new])

    except:
        df=pd.concat([df,df2])
```

```python
df.rename(columns={'Code': 'Code-EODapi', 'CUSIP': 'CUSIP-EODapi', 'Sector': 'Sector-EODapi',
                  'Industry': 'Industry-EODapi', 'GicSector': 'GicSector-EODapi', 'GicGroup': 'GicGroup-EODapi',
                  'GicIndustry': 'GicIndustry-EODapi', 'GicSubIndustry': 'GicSubIndustry-EODapi'}, inplace=True)

df = df.dropna(subset=['ISIN'])

df
```
