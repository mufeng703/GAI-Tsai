```python
import numpy as np  
import pandas as pd 
```

```python
path = "./Data/withFIGI.csv"

df = pd.read_csv(path, index_col=0)  
df
```

```python
df = df.drop(['Not_Match', 'Source_Available'], axis=1)

df
```

Country, Sector/Industry, Exchange, ISIN, FIGI

```python
df_ck = df.copy()
df_ck['Ranking'] = 0

columns = ['Country-EOD', 'Exchange-EOD', 'ISIN', 'country-Inv', 
           'figi', 'exchCode-FIGI']

for col in columns:
    df_ck.loc[~df_ck[col].isnull(),'Ranking'] = 1 + df_ck['Ranking'].astype(int)

df_ck[['Name_197','Ranking']]
```

```python
df_ck['Ranking'].value_counts().to_frame()
```

## Change Stock Type to Numer

```python
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
df_ck['stockstockType_197_Num'] = LE.fit_transform(df_ck['stockType_197'])

df_ck[['stockType_197', 'stockstockType_197_Num']]
```

```python
df_ck['stockstockType_197_Num'].value_counts().to_frame()
```
