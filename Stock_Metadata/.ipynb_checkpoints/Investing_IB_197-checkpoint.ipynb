```python
import numpy as np
import pandas as pd
```

```python
path1 = "./Data/full stock list.csv"

df_inv = pd.read_csv(path1)  
df_inv = df_inv.rename(columns={"Ticker": "Ticker-Inv"})

df_inv.head()
```

```python
path = "./Data/ISIN_from_IB.csv"

df_IB = pd.read_csv(path)  
df_IB
```

```python
df_IB = df_IB.assign(**{'ISIN-IB': df_IB['ISIN-IB'].fillna(df_IB['ISIN-19.7'])})
df_IB.drop(['ISIN-19.7'], inplace=True, axis=1)
df_IB.dropna(subset=['ISIN-IB'],inplace=True)
df_IB
```

```python
len(df_IB['ISIN-IB'].unique())
```

```python
df_IB = df_IB.rename(columns={"ISIN-IB": "ISIN", "Ticker": "Ticker-IB"})
df_IB
```

```python
df = pd.merge(df_IB, df_inv, how='outer', on='ISIN')
df
```

```python
df = df.rename(columns={"Name": "Name-Inv", "Exchange": "Exchange-Inv", "Type": "Type-Inv", "Market": "Market-Inv", "CUSIP": "CUSIP-Inv"})
df
```
