```python
path1 = "./full stock list.csv"

df_inv = pd.read_csv(path1)  
df_inv.head()
```

```python
path = "/Users/caichengyun/Documents/User/Documents/Master/Intern/Global AI/Projects/Stock_Metadata/ISIN_from_IB.csv"

df_IB = pd.read_csv(path)  
df_IB
```

```python
df_IB = df_IB.assign(**{'ISIN-IB': df_IB['ISIN-IB'].fillna(df_IB['ISIN-19.7'])})
df_IB
```

```python
df = pd.merge(df_IB, df_inv, how='outer', on='Ticker')
df
```

```python
df = df.assign(**{'ISIN': df['ISIN'].fillna(df['ISIN-IB'])})
df
```
