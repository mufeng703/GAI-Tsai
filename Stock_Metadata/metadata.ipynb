```python
import numpy as np
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
path1 = "./Data/withEOD.csv"

df_eod = pd.read_csv(path1)  
df_eod.head()
```

```python
path = "./Data/LIst of 19.7 Stocks with FIGI if available - By Evelyn.xlsx"

df_stockList = pd.read_excel(path, index_col=0)  
df_stockList.head()
```

```python
df = df_eod[['EOD-Code', 'EOD-Name', 'Symbol-IB', 'longName-IB' ]]
df = df.assign(**{'EOD-Code': df['EOD-Code'].fillna(df['Symbol-IB'])})
df = df.assign(**{'EOD-Name': df['EOD-Name'].fillna(df['longName-IB'])})
df = df.dropna()
df
```

```python
df_197 = df_stockList[['symbol_x', 'longName']]
df_197
```

```python
match = []
for index, row in df_197.iterrows():
    
    if row['symbol_x'] in df['EOD-Code'].values :
        df_sym = df[df['EOD-Code'] == row['symbol_x']].copy()
        for index2, row2 in df_sym.iterrows():
            fuzz_socre1 = fuzz.token_sort_ratio(row['longName'],row2['EOD-Name'])
            fuzz_socre2 = fuzz.token_sort_ratio(row['longName'],row2['longName-IB'])
            
            if fuzz_socre1>90 or fuzz_socre2>90:
                row2['Name_197'] = row['longName']
                match.append(row2)

                if len(match) % (len(df_197)//10) == 0:
                    print("progress: %.2f" % (100*len(match)/len(df_197)) + "%")
    
    elif row['symbol_x'] in df['Symbol-IB'].values :
        df_sym = df[df['Symbol-IB'] == row['symbol_x']].copy()
        for index2, row2 in df_sym.iterrows():
            fuzz_socre1 = fuzz.token_sort_ratio(row['longName'],row2['EOD-Name'])
            fuzz_socre2 = fuzz.token_sort_ratio(row['longName'],row2['longName-IB'])
            
            if fuzz_socre1>90 or fuzz_socre2>90:
                row2['Name_197'] = row['longName']
                match.append(row2)

                if len(match) % (len(df_197)//10) == 0:
                    print("progress: %.2f" % (100*len(match)/len(df_197)) + "%")
    
```

```python
df_match = pd.DataFrame(match)
df_match
```

```python
df_match.to_csv('./Data/metaDataFinal.csv', index=False)
```
