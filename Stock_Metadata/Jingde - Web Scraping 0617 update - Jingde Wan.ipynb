```python
base_url = "https://www.investing.com"
titles = ['Name', 'Type', 'Market', 'ISIN', 'CUSIP']
official_names = []
names = []
types = []
markets = []
isins = []
cusips = []
exchanges = []
for url in stockURL:
    page  = requests.get(base_url + str(url.select("a")[0]['href']))
    print(str(url.select("a")[0]['href']))
    soupStock = BeautifulSoup(page.content, 'html.parser')
    full_information = soupStock.find(class_='instrument-header_instrument-header__1SRl8 mb-5 bg-background-surface tablet:grid tablet:grid-cols-2')
    
    # find exchange
#     if (soupStock.find(class_='text-xs ml-1') is not None):
#         print(type(soupStock.find(class_='text-xs ml-1')))
#         exchange = soupStock.find(class_='text-xs ml-1').get_text()
#     if (soupStock.find(class_='text-xs font-semibold leading-tight ml-2.5') is not None):
#         print(type(soupStock.find(class_='text-xs font-semibold leading-tight ml-2.5')))
#         exchange = soupStock.find(class_='text-xs font-semibold leading-tight ml-2.5').get_text()
#     else:
#         exchange = 'NA'
    
    try:
        try:
            print('text-xs ml-1')
            print(type(soupStock.find(class_='text-xs ml-1')))
            exchange = soupStock.find(class_='text-xs ml-1').get_text()
        except:
            print('text-xs font-semibold leading-tight ml-2.5')
            print(soupStock.find(class_='text-xs font-semibold leading-tight ml-2.5'))
            exchange = soupStock.find(class_='text-xs font-semibold leading-tight ml-2.5').get_text()
    except:
        print('None Type')
        continue
        
    full_name = full_information.find('h1', attrs={'class': "text-2xl font-semibold instrument-header_title__GTWDv mobile:mb-2"}).get_text()
    name = re.findall(r'\(.*?\)', full_name)[0].replace("(","").replace(")","")
    official_name = full_name.split("(")[0].strip() 
#     print(name)
    
    information = full_information.find_all(class_='bottom-bar_value__1F-pW')
    info = []
    for i in information:
        info.append(i.get_text())
    info.insert(0, name)
    official_names.append(official_name)
    names.append(info[0])
    types.append(info[1])
    markets.append(info[2])
    isins.append(info[3])
    try:
        cusips.append(info[4])
    except:
        cusips.append('NA')
    exchanges.append(exchange)
```
