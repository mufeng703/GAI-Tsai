```python
import numpy as np
import pandas as pd
import requests 
```

```python
path = "./Data/metaData_IB_Inv.csv"

df_isin = pd.read_csv(path)  
df_isin
```

```python
df_notNull = df_isin[~df_isin['ISIN'].isnull()].copy()
df_notNull
```

```python
df_eod = pd.DataFrame()
for isin in df_notNull["ISIN"]:
    if str(isin) != 'nan':
        path = "https://eodhistoricaldata.com/api/search/"+str(isin)+"?api_token=5f3afd582bd7b4.95720069"
        
        df = pd.read_json(path) 
        if len(df) != 0:
            df_eod = df_eod.append(df.iloc[0], ignore_index=True)
            
    if len(df_eod) % (len(df_notNull)//5) == 0:
        print("progress: %.2f" % (100*len(df_eod)/len(df_notNull)) + "%")
```

```python
df_eod
```

```python
data = pd.merge(df_eod, df_isin, how='outer', on='ISIN')
data
```

```python
data = data.rename(columns={"Code": "Code-EOD", "Country": "Country-EOD", "Currency": "Currency-EOD", "Exchange": "Exchange-EOD", "Name": "Name-EOD" , "Type": "Type-EOD" , "previousClose": "previousClose-EOD" , "previousCloseDate": "previousCloseDate-EOD"})
data
```

```python
data = data.drop_duplicates(subset=['ISIN'])

data
```

```python
data.to_csv('withEOD.csv', index=False)
```
